```python
def load_network_from_csv(url, cut_after):
    raw_df = pd.read_csv(url,compression='gzip')
    if cut_after is not None:
        raw_df = raw_df[raw_df['ending_times'] < cut_after]
        
    source_nodes = raw_df['source_nodes'].tolist()
    target_nodes = raw_df['target_nodes'].tolist()
    starting_times = raw_df['starting_times'].tolist()
    ending_times = raw_df['ending_times'].tolist()
    return source_nodes, target_nodes,starting_times, ending_times
```

```python
url = 'https://zenodo.org/record/4725155/files/'\
                        'mice_contact_sequence.csv.gz'
cut_after = 24*3600 # only use first day
source_nodes, target_nodes,starting_times, ending_times=load_network_from_csv(url, cut_after=None)
```

```python
cont_net=tn.ContTempNetwork(\
    source_nodes=source_nodes,\
    target_nodes=target_nodes,\
    starting_times=starting_times,\
    ending_times=ending_times\
)
```
